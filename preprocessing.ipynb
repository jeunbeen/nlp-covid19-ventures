```python
import pandas as pd
```

```python
data = pd.read_csv('한국경제 크롤링.csv', encoding = 'cp949')
```

```python
text = data.body
text
```

```python
# 문장으로 분리

import kss

sentence_tokenized_text = []
for i, line in enumerate(text):
    line = line.strip()
    for sent in kss.split_sentences(line):
        sentence_tokenized_text.append(sent.strip())
        
sentence_tokenized_text[:2]
```

```python
punct = "/-'?!.,#$%\'()*+-/:;<=>@[\\]^_`{|}~" + '""“”’' + '∞θ÷α•à−β∅³π‘₹´°£€\×™√²—–&'
punct_mapping = {"‘": "'", "₹": "e", "´": "'", "°": "", "€": "e", "™": "tm", "√": " sqrt ", "×": "x", "²": "2", "—": "-", "–": "-", "’": "'", "_": "-", "`": "'", '“': '"', '”': '"', '“': '"', "£": "e", '∞': 'infinity', 'θ': 'theta', '÷': '/', 'α': 'alpha', '•': '.', 'à': 'a', '−': '-', 'β': 'beta', '∅': '', '³': '3', 'π': 'pi', }

def clean_punc(text, punct, mapping): 
    for p in mapping: 
        text = text.replace(p, mapping[p])

    for p in punct: 
        text = text.replace(p, f' {p} ')

    specials = {'\u200b': ' ', '…': ' ... ', '\ufeff': '', 'करना': '', 'है': ''} 
    for s in specials: 
        text = text.replace(s, specials[s])

    return text.strip()

no_punc = []
for line in text:
    no_punc.append(clean_punc(line, punct, punct_mapping))
no_punc[:2]
```
